```python
# Homework: Find an arbitrary document and tag it using a tagger. 
# Store the tagged file and draw frequency distribution of POS tags in it.
from nltk import word_tokenize
with open ('doc.txt') as fin, open('tokens.txt','w') as fileout:
    for line in fin:
        tokens = word_tokenize(line)
        print(' '.join(tokens), end=' ', file=fileout)
with open ("tokens.txt", "r") as myfile:
    data=myfile.readlines()
print('My document content:\n',data)
data_tokenize= word_tokenize(data[0])
print("\n\n\n\nReading tagged document:\n\n\n")
print(nltk.pos_tag(data_tokenize, tagset='universal'))
doc_tagged= nltk.pos_tag(data_tokenize, tagset='universal')
```

```python
tag_fd = nltk.FreqDist(tag for (word, tag) in doc_tagged)
print(tag_fd.most_common())
tag_fd.plot(cumulative=True)
```

```python
wsj = nltk.corpus.treebank.tagged_words(tagset='universal')
wsj = nltk.corpus.treebank.tagged_words()
cfd1 = nltk.ConditionalFreqDist(wsj)
cfd2 = nltk.ConditionalFreqDist((tag, word) for (word, tag) in wsj)

print(cfd1['play'].most_common())
```
